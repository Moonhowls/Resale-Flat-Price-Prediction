## Question: Can we predict the price of resale flats?

In the resale flat market, homeowners may want to know the prices that their flats will fetch if they choose to sell them. On the other side, homebuyers would want to know what is a fair price for a resale flat.

So, is it possible to predict the price of a resale flat accurately using machine learning techniques? If yes, how accurate will the predictions be?

Let's find out.

## Approach 1: Bi-variate Linear Regression

First, we will use bi-variate linear regression to predict the resale flat prices. 

There is only one other numeric variable besides resale price, which is floor area. Thus, we will use floor area to predict resale price using a linear regression model.

### Using Floor Area to Predict Resale Price

### Step 1: Plot floor area against resale price.

```python
sb.jointplot(data = RFP, x = "floor_area_sqm", y = "resale_price", height = 12)
```

### Step 2: Extract resale price and floor area into a single dataframe.

```python
RP = pd.DataFrame(RFP['resale_price'])
FA = pd.DataFrame(RFP['floor_area_sqm'])

FARP = pd.concat([RP,FA], axis = 1)
FARP
```

### Step 3: Take a look at the correlation coefficients between floor area and resale price.

```python
sb.heatmap(FARP.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```

The coefficient is relatively high, meaning that there may be a strong linear relationship between floor area and resale flat prices.

### Step 4: Spliting Data into Train and Test Data

We will use 67593 rows as our train data and 22530 rows as our test data. (~75% train, ~25% test)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

RP_train, RP_test, FA_train, FA_test = train_test_split(RP , FA , test_size = 0.249991678)

print(RP_train)
print(RP_test)
```

### Step 5: Fitting Linear Regression Model Onto Train Data

So now we want to fit a linear regression model onto the data.

Regression Problem: Resale_Price = a X Floor_Area + b

#### Thus, resale price is your response and floor area is your predictor.

```python
#Create a Linear Regression object
linreg = LinearRegression()

#Train the Linear Regression object
linreg.fit(FA_train , RP_train)

print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_)
```

```python
RP_train_pred = linreg.predict(FA_train)

f = plt.figure(figsize=(16, 8))
plt.scatter(FA_train , RP_train)
plt.scatter(FA_train , RP_train_pred, color = "r")
plt.show()
```

```python
print("Explained Variance (R^2) \t:", linreg.score(FA_train , RP_train))
```

```python
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))

train_mse = mean_sq_err(FA_train , RP_train)
train_rmse = np.sqrt(train_mse)
print("Mean Squared Error (MSE) \t:", train_mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(train_mse))
```

### Step 6: Apply the Model onto Test Data

Now, we apply the model on the test data.

```python
RP_test_pred = linreg.predict(FA_test)

f = plt.figure(figsize=(16, 8))
plt.scatter(FA_test,RP_test , color = "green")
plt.scatter(FA_test, RP_test_pred, color = "red")
plt.show()
```

```python
print("Explained Variance (R^2) \t:", linreg.score(FA_test, RP_test))

# Mean Squared Error (MSE)
def mean_sq_err(actual, predicted):
    '''Returns the Mean Squared Error of actual and predicted values'''
    return np.mean(np.square(np.array(actual) - np.array(predicted)))

test_mse = mean_sq_err(RP_test, RP_test_pred)
test_rmse = np.sqrt(test_mse)
print("Mean Squared Error (MSE) \t:", test_mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(test_mse))
```

## Analysis of Linear Regression Model to Predict Resale Price using Floor Area

### 1. Explained Variance

```python
print("Explained Variance of Model on Train Data: \t", linreg.score(FA_train , RP_train))
print("Explained Variance of Model on Test Data: \t", linreg.score(FA_test, RP_test))
```

### 2. Mean Squared Error

```python
print("Mean Squared Error of Model on Train Data: \t", train_mse)
print("Mean Squared Error of Model on Test Data: \t", test_mse)
```

### 3. Root Mean Squared Error

```python
print("Root Mean Squared Error of Model on Train Data: \t", train_rmse)
print("Root Mean Squared Error of Model on Test Data: \t", test_rmse)
```

### Overall Analysis for Bi-Variate Linear Regression Model

The explained variance is not very close to 1, and the mean squared errors and root mean squared errors are quite significant, thus predicting resale price using floor area and the linear regression model may not be very accurate, and may not be the best approach.

## Approach 2: Multi-variate Regression Tree

How about we try another approach? We can make use of the different variables in the data and apply a decision tree model to predict resale price.

Let's use the town, flat type, storey range, floor area and flat model variables. We choose not to use the lease commencement year and remaining lease as we want to see if we can predict the price of the flat without knowing any information about the lease.

Let's take a look at these few variables.

```python
X = pd.DataFrame(RFP[["town","flat_type","storey_range", "floor_area_sqm", "flat_model"]])
X
```

### Step 1: One-Hot Encoding of Categorical Predictors

It can be seen that there are a mix of categorical and numeric variables in our predictors. However, the scikit decision tree models do not support categorical data in their original form. They must be converted to one-hot encoded form.

```python
town_1HE = pd.get_dummies(RFP.town)
print(town_1HE)
```

```python
FT_1HE = pd.get_dummies(RFP.flat_type)
print(FT_1HE)
```

```python
FM_1HE = pd.get_dummies(RFP.flat_model)
print(FM_1HE)
```

```python
SR_1HE = pd.get_dummies(RFP.storey_range)
print(SR_1HE)
```

Here we concatenate the different one-hot encoded data to a single dataframe called X_1HE. X_1HE will be our predictor in the decision tree model.

```python
floorarea = pd.DataFrame(RFP['floor_area_sqm'])

X_1HE = pd.concat([town_1HE, FT_1HE, SR_1HE, floorarea, FM_1HE],axis = 1)
print(X_1HE)
```

### Step 2: Splitting of Data into Train and Test Data

We will use 67593 rows as our train data and 22530 rows as our test data. (~75% train, ~25% test)

```python
RP_train, RP_test, X_train, X_test = train_test_split(RP ,X_1HE, test_size = 0.249991678)
print(RP_train)
print(X_train)
```

### Step 3: Train the Regression Tree Model using the Train Data

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import OneHotEncoder
```

```python
# Decision Tree using Train Data
dectree = DecisionTreeRegressor()  # create the decision tree object
dectree.fit(X_train, RP_train)                    # train the decision tree model

# Predict Response corresponding to Predictors
RP_train_pred = dectree.predict(X_train)
RP_test_pred = dectree.predict(X_test)
```

```python
print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", dectree.score(X_train, RP_train))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", dectree.score(X_test, RP_test))
print()
```

```python
from sklearn.metrics import mean_squared_error as MSE

#Compute test set MSE
dectree_mse = MSE(RP_test, RP_test_pred)

#Compute test set RMSE
dectree_rmse = dectree_mse**(1/2)

print("Mean Squared Error of model on Test Data:\t", dectree_mse)
print("Root Mean Squared Error of model on Test Data:\t", dectree_rmse)
```

## Analysis of Multi-Variate Regression Tree

The classification accuracy of this model is very high, which means that it is possible to predict resale prices using a multi-variate classification tree model.

The mean-squared error and root mean-squared error of the model is also significantly lower than the errors of the bi-variate linear regression model, which implies that this model is better than the bi-variate linear regression model.

Is there a way we can make the results possibly better?

Yes, we can use a Random Forest model.

## Approach 3: Random Forest

A random forest is an approach that uses multiple decision trees with randomly selected datasets and variables used for each decision tree. 

### Step 1: Preparing Data

We convert our data into numpy arrays, then split them up into train and test data.

```python
# Labels are the values we want to predict
RP_labels = np.array(RFP['resale_price'])

# Saving feature names for later use
feature_list = list(X_1HE.columns)

# Convert to numpy array
X_features = np.array(X_1HE)
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(X_features ,RP_labels, test_size = 0.249991678)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)

print(train_labels)
```

### Step 2: Train the Model

We need to train our random forest model with the train data.

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor

# Instantiate model
rf = RandomForestRegressor()

# Train the model on training data
rf.fit(train_features, train_labels)
```

### Step 3: Apply the model to the Test Data

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'dollars')
```

```python
print("Goodness of Fit of Model \tTrain Dataset")
print("Classification Accuracy \t:", rf.score(train_features, train_labels))
print()

# Check the Goodness of Fit (on Test Data)
print("Goodness of Fit of Model \tTest Dataset")
print("Classification Accuracy \t:", rf.score(test_features, test_labels))
print()
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

## Analysis of Random Forest Model

The random forest model has predicted the resale flat prices with a very high accuracy of around and above 93%. 

## Conclusion

Out of all the models, it can be seen that the random forest model and the multi-variate regression tree model can be used to predict resale flat prices with a high accuracy using multiple variables.

Under normal circumstances, we will expect the random forest model to yield a much better result than the multi-variate regression tree, however, this dataset works well even with only 1 decision tree fitted to it.

Thus, it is not necessary to use the random forest model to predict resale flat prices as the multi-variate regression tree model has a very high classification accuracy and relatively low errors.

Therefore, we can predict resale flat prices very accurately, with more than 90% accuracy, by using the multi-variate regression tree model or the random forest model.
